```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
print(iris.data)
```

```python
print(iris.target)
```

```python
from sklearn import svm
clf=svm.SVC(gamma=0.001, C=100.)
```

```python
clf.fit(iris.data,iris.target)
```

```python
clf.predict([[5.1,3.5,1.4,0.2]])
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=2000)
```

```python
mlp.fit(iris.data,iris.target)
```

```python
mlp.predict([[5.1,3.5,1.4,0.2]])
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions=mlp.predict(iris.data)
```

```python
print(confusion_matrix(iris.target,predictions))
```
